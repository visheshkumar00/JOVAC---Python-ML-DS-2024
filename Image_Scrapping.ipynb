```python
import requests
import bs4
```

```python
url = 'https://www.snapdeal.com/search?keyword=iphone&santizedKeyword=&catId=&categoryId=0&suggested=false&vertical=&noOfResults=20&searchState=&clickSrc=go_header&lastKeyword=&prodCatId=&changeBackToAll=false&foundInAll=false&categoryIdSearched=&cityPageUrl=&categoryUrl=&url=&utmContent=&dealDetail=&sort=rlvncy'
```

```python
response = requests.get(url)
```

```python
print(response)
```

```python
response.content
```

```python
soup = bs4.BeautifulSoup(response.content)
```

```python
print(soup)
```

```python
first_phto = soup.find('img')
```

```python
print(first_phto)
```

```python
print(first_phto.src)
```

```python
all_photos = soup.findAll('img')
```

```python
all_photos
```

```python
img = all_photos[0]
```

```python
img
```

```python
all_tags = img.attrs
```

```python
all_tags
```

```python
all_tags['data-src']
```

```python
with open('scrapped_img.jpeg', 'wb') as file:
    image = requests.get(all_tags['data-src']).content
    file.write(image)
```
