```python
table = data_all.find("table", {'class': 'wikitable sortable sticky-header col2left'})
```

```python
table
```

```python
print(type(table))
```

```python
heading = table.find_all('th')
```

```python
cols = []

for i in heading:
    cols.append(i.text[:-1])
```

```python
cols
```

```python
all_data_table = table.find_all('tr')
```

```python
table_row = []

for i in all_data_table:
    curr_row = []
    row = i.find_all('td')
    
    for col_data in row:
        curr_row.append(col_data.text)
    
    table_row.append(curr_row)
```

```python
print(table_row)
```

```python
file_name = 'C:/Users/bhatt/OneDrive/Desktop/whisper-main/DS_GLA_Jovac/scrapped_table.csv'
```

```python
with open(file_name, 'w', encoding = 'utf_8') as file:
    heading = ','.join(cols)
    heading += '\n'
    
    file.write(heading)
    
    for row in table_row:
        temp = ''
        for ele in row:
            temp += ele.replace(',', '.')
            temp += ','
        print(temp[:-1])
        file.write(temp[:-1])
```
